```python
import pandas as pd
from scipy import stats
import numpy as np

def explanatory_analysis(charges_data_path, personal_data_path, plan_data_path):
    # write you solution here
    #read all datasets
    charges = pd.read_csv(charges_data_path)
    personal = pd.read_csv(personal_data_path)
    plan = pd.read_csv(plan_data_path)

    #get the monthly charges mean
    all_charges_values = charges['monthlyCharges'].values
    observed_values = all_charges_values[~np.isnan(all_charges_values)]
    monthly_charges_mean = int(round(stats.trim_mean(observed_values, 0.1)))

    #fill missing values in monthlyCharges and totalCharges
    charges['monthlyCharges'] = charges['monthlyCharges'].fillna(monthly_charges_mean)
    charges['totalCharges'] = charges['totalCharges'].fillna(monthly_charges_mean*charges['tenure'])

    #create tenurebinned
    charges['tenureBinned']  = pd.cut(charges['tenure'], bins=[0, 24,48,60,float("inf")], labels=['group1', 'group2', 'group3', 'group4'])
    charges_date_updated = charges

    # calculate churn rate
    churn_rate = len(charges.loc[charges['churn'] == 'Yes'])/len(charges)
    churn_pct = int(round(churn_rate,2)*100)

    # join
    charges_personal = pd.merge(charges, personal, on='customerID')
    data_merged = pd.merge(charges_personal, plan, how='left',on='customerID')

    # age above 60
    customer_age = data_merged[['customerID', 'age']].drop_duplicates()
    age_above_60 = len(customer_age.loc[customer_age['age'] > 60 ])/len(customer_age)
    pct_age_above_60 = int(round(age_above_60,2)*100)

    # dict
    internet_service_counts = data_merged.groupby('internetService')['customerID'].nunique().to_dict()


    results = {
        monthly_charges_mean,
        charges_date_updated,
        churn_pct,
        data_merged,
        pct_age_above_60,
        internet_service_counts
    }
    return results
```
